```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dims = 30
V = np.arange(0, 1e3, 10)
E = dims * np.ones(len(V))
```

```python
vv = np.random.randint(-5e4, 8e4, size=V.shape)
ve = np.random.randint(-5e4, 8e4, size=V.shape)
```

```python
plt.plot(V**2 + vv, 'r')
plt.plot(V*E + ve, 'k')
plt.plot(V**2, 'k--')
plt.plot(V*E, 'r--')
plt.show()
```
